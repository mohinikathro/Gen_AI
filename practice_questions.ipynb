```python
## 1) create a new list containing only numbers divisible by 6.
nums = [3, 6, 9, 12, 15, 18]
nums_6 = []

for num in nums:
    if num %6==0:
        nums_6.append(num)

print(nums_6)

## 2) Create a list of words converted to lowercase where the word length is greater than 4.
words = ['Python', 'data', 'Science', 'AI', 'learning']

lst = []
for word in words:
    if len(word) >4:
        lst.append(word.lower())

print(lst)

## 3) Create a dictionary that stores name → count of occurrences.
names = ['Alice', 'Bob', 'Charlie', 'Bob', 'Alice']

count = {}
for name in names:
    count[name] = count.get(name, 0) + 1

print(count)

## 4) Create a dictionary of subjects where the score is ≥ 85.
scores = {
    'math': 82,
    'science': 90,
    'english': 70,
    'history': 88
}

subjects = {}

for subject, score in scores.items():
    if score >= 85:
        subjects[subject] = score

print(subjects)

## 5) Find the common elements using sets.
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

common = set(list1).intersection(set(list2))
print(common)

## 6) Write a function count_even(nums) that returns the number of even elements in a list.

nums = [2,12,34,1234,566,3247, 889, 47, 227, 359, 3389, 226, 28289, 23785]

def count_even(nums):
    count = 0
    for num in nums:
        if num%2==0:
            count += 1

    return count

print(count_even(nums))

## 7) Write a function word_lengths(words) that returns a dictionary where word → length of word.
words = ['Python', 'data', 'Science', 'AI', 'learning']


def word_leng(words):
    dict4 = {}
    for word in words:
        dict4[word] = len(word)
    return dict4

print(word_leng(words))


## 8) Write a lambda function that checks whether a number is positive, negative, or zero.

number = -3

check = (lambda x:  'positive' if x>0 else 'negative' if x<0 else 'zero')
print(check(number))



## 9) return a list where each number is multiplied by 3.
nums = [1, 2, 3, 4, 5]

res = list(map(lambda x: x*3, nums))
print(res)


## 10) 

students = [
    {'name': 'Alice', 'marks': 80},
    {'name': 'Bob', 'marks': 65},
    {'name': 'Charlie', 'marks': 90},
    {'name': 'David', 'marks': 70}
]

res1 = {
    'passed': [],
    'failed': []
}

for student in students:
    if student['marks'] >=75:
        res1['passed'].append(student['name'])
    else:
        res1['failed'].append(student['name'])

print(res1)

```
